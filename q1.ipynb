```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colleges = pd.read_csv("MergerdAyush_Colleges_in_India_01-04-2021_0.csv")
colleges.head()
```

```python
colleges.columns
```

```python
colleges = colleges.fillna(0)
```

```python
new_data = colleges.groupby('State or Union Territory')
govt_colleges = new_data['No. of Colleges - Govt'].sum()
non_govt_colleges = new_data['No. of Colleges - Non-Govt'].sum()
result = pd.DataFrame({'Government Colleges': govt_colleges, 'Non-Government Colleges': non_govt_colleges})
result.reset_index(inplace=True)
result
```

```python
types = ["screen","poster print","presentation"]
color_pallete = ["rocket","Purples","BrBG"]
for i in types:
    if i == "screen":
        sns.set_style('whitegrid')
        fig,axs = plt.subplots(1,4,figsize=(20,10))
        fig.suptitle("Screen Display")
        for j in range(3):
            sns.set_palette(color_pallete[j])
            sns.set_context("notebook", rc={"grid.linewidth": 0.3, 'lines.markersize': 6, 'lines.linewidth': 1.5})
            result.plot(kind='barh',stacked=True,ax=axs[j])
            axs[j].set_title(color_pallete[j])
        sns.set_palette(['Red', 'Green'])
        sns.set_context("notebook", rc={"grid.linewidth": 0.3, 'lines.markersize': 6, 'lines.linewidth': 1.5})
        result.plot(kind='barh',stacked=True,ax=axs[3])
        axs[3].set_title("Custom")
    if i == "poster print":
        sns.set_style('whitegrid')
        fig,axs = plt.subplots(1,4,figsize=(20,10))
        fig.suptitle("Poster Print")
        for j in range(3):
            sns.set_palette(color_pallete[j])
            sns.set_context("poster", rc = {"grid.linewidth": 0.5})
            result.plot(kind='barh',stacked=True,ax=axs[j],legend=False)
            axs[j].set_title(color_pallete[j])
        sns.set_palette(['Red', 'Green'])
        sns.set_context("poster", rc = {"grid.linewidth": 0.5})
        result.plot(kind='barh',stacked=True,ax=axs[3],legend=False)
        axs[3].set_title("Custom")
    if i == "presentation":
        sns.set_style('whitegrid')
        fig,axs = plt.subplots(1,4,figsize=(20,10))
        fig.suptitle("Presentation")
        for j in range(3):
            sns.set_palette(color_pallete[j])
            sns.set_context("talk")
            result.plot(kind='barh',stacked=True,ax=axs[j])
            axs[j].set_title(color_pallete[j])
        sns.set_palette(['Red', 'Green'])
        sns.set_context("talk")
        result.plot(kind='barh',stacked=True,ax=axs[3])
        axs[3].set_title("Custom")
```

```python
agriculture = pd.read_csv("Production_of_Agriculture_Stock-Ministry_of_Finance.csv")
agriculture= agriculture.drop(["Frequency","Unit"],axis=1)
agriculture= agriculture.drop(index=0)
agriculture.loc[1,"Particulars"] = "Rice"
agriculture.loc[2,"Particulars"] = "Paddy"
agriculture.loc[3,"Particulars"] = "Wheat"
agriculture.head()
```

```python
agriculture_new = agriculture.melt(id_vars=['Particulars'], var_name='Month', value_name='Production')
agriculture_new.head()
```

```python
types = ["screen","poster print","presentation"]
color_pallete = ["rocket","Purples","BrBG"]
for i in types:
    if i == "screen":
        sns.set_style('whitegrid')
        fig,axs = plt.subplots(1,4,figsize=(20,10))
        fig.suptitle("Screen Display")
        for j in range(3):
            sns.set_palette(color_pallete[j])
            sns.set_context("paper", rc={"grid.linewidth": 0.3, 'lines.markersize': 6, 'lines.linewidth': 1.5})
            sns.lineplot(x='Month', y='Production', hue='Particulars', style="Particulars", dashes=False,data=agriculture_new,ax=axs[j])
            axs[j].set_title(color_pallete[j])
            axs[j].tick_params(axis='x', labelrotation=90)
        sns.set_palette(['Red', 'Green','Blue'])
        sns.set_context("paper", rc={"grid.linewidth": 0.3, 'lines.markersize': 6, 'lines.linewidth': 1.5})
        sns.lineplot(x='Month', y='Production', hue='Particulars', style="Particulars", dashes=False,data=agriculture_new,ax=axs[3])
        axs[3].set_title("Custom")
        axs[3].tick_params(axis='x', labelrotation=90)
    if i == "poster print":
        sns.set_style('whitegrid')
        fig,axs = plt.subplots(1,4,figsize=(20,10))
        fig.suptitle("Poster Print")
        for j in range(3):
            sns.set_palette(color_pallete[j])
            sns.set_context("poster", rc = {"grid.linewidth": 0.5})
            sns.lineplot(x='Month', y='Production', hue='Particulars', style="Particulars", dashes=False,data=agriculture_new,ax=axs[j])
            axs[j].set_title(color_pallete[j])
            axs[j].tick_params(axis='x', labelrotation=90)
        sns.set_palette(['Red', 'Green','Blue'])
        sns.set_context("poster", rc = {"grid.linewidth": 0.5})
        sns.lineplot(x='Month', y='Production', hue='Particulars', style="Particulars", dashes=False,data=agriculture_new,ax=axs[3])
        axs[3].set_title("Custom")
        axs[3].tick_params(axis='x', labelrotation=90)
    if i == "presentation":
        sns.set_style('whitegrid')
        fig,axs = plt.subplots(1,4,figsize=(20,10))
        fig.suptitle("Presentation")
        for j in range(3):
            sns.set_palette(color_pallete[j])
            sns.set_context("talk")
            sns.lineplot(x='Month', y='Production', hue='Particulars', style="Particulars", dashes=False,data=agriculture_new,ax=axs[j])
            axs[j].set_title(color_pallete[j])
            axs[j].tick_params(axis='x', labelrotation=90)
        sns.set_palette(['Red', 'Green','Blue'])
        sns.set_context("talk")
        sns.lineplot(x='Month', y='Production', hue='Particulars', style="Particulars", dashes=False,data=agriculture_new,ax=axs[3])
        axs[3].set_title("Custom")
        axs[3].tick_params(axis='x', labelrotation=90)
```

```python
transport = pd.read_csv("Road_Transport_BRS_2017-18_Annexure7_9c.csv")
transport.head()
```

```python
transport_data = pd.pivot_table(transport, index=['Name of the States'], values=['National Highways','State Highways','District Roads'])
```

```python
transport_data.reset_index(inplace=True)
transport_data
```
